```python
subplot
 subplot(X,Y,index) - index => 1 2 3 4 ..N
 subplot(2,1,1) ----same as ---> subplot(211)
 subplot(2,1,2) ----same as ---> subplot(212)
```

```python
X = np.arange(0,5,0.1)
Y = np.sin(2*np.pi*d)
plt.subplot(2,1,1)
plt.plot(X,Y,'ro--')
plt.title('Abnormal values')

X = np.arange(0,5,0.1)
Y = np.sin(2*np.pi*d)
plt.subplot(2,1,2)
plt.plot(X,Y,'bs--')
plt.title('Normal values')
```
