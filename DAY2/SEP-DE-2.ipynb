```python
from scipy.stats import zscore


df_IP = pd.DataFrame({'Key1':[10,20,30,40,50,600]})
Q1 = df_IP['Key1'].quantile(0.25)
Q3 = df_IP['Key1'].quantile(0.75)
IQR = Q3-Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
#df_IP['Key1'] >= lower_bound and df_IP['Key1'] <= upper_bound 
```

```python
def fx_remove_outliers(df,column,threshold=3):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df_IP = pd.DataFrame({'Key1':[10,20,30,40,50,600]})
filtered_result = fx_remove_outliers(df_IP,'Key1')
filtered_result
```

```python
df_IP2 = pd.DataFrame({'Key1':[10,20,300,40,50],'Key2':[11,12,13,14,15]})
filtered_result = fx_remove_outliers(df_IP2,'Key1')
filtered_result
```

```python
df_IP3 = pd.DataFrame({'Key1':[10,20,300,40,50],'Key2':[11,12,130,14,15]})
filtered_result = fx_remove_outliers(df_IP3,'Key2')
filtered_result
```
